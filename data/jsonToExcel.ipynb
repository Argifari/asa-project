```python
import pandas as pd
import json

with open('solData.json', 'r') as f:
    jsonData = json.load(f)

df = pd.DataFrame(jsonData)

kolom_baru = {
    "move_id": "ID",
    "move_name": "Nama Gerakan",
    "move_notation": "Notasi",
    "damage": "Damage",
    "startup": "Startup Frame",
    "active": "Active Frame",
    "recovery": "Recovery Frame",
    "attack_level": "Attack Level",
    "hitstun": "Hitstun",
    "proration": "Proration"
}


df = df.rename(columns=kolom_baru)


outputFile = "SolFrameData.xlsx"
df.to_excel(outputFile, index=False)

print(f"[SUKSES] Berkas spreadsheet berhasil dibuat dengan nama : {outputFile}")
print("[SELESAI]")
```
